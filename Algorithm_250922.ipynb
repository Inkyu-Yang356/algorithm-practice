```python
def find_max(a):
    n = len(a)
    max_v = a[0]
    for i in range(1, n):
        if a[i] > max_v:
            max_v = a[i]
    return max_v

v = [17, 92, 18, 33, 58, 7, 33, 42]
print(find_max(v))
```

```python
def find_max_idx(a):
    n = len(a)
    max_idx = 0
    for i in range(1, n):
        if a[i] > a[max_idx]:
            max_idx = i
    return max_idx

v = [17, 92, 18, 33, 58, 7, 33, 42]
print(find_max_idx(v), v[find_max_idx(v)])
```

```python
def find_same_name_set(a):
    n = len(a)
    result = set()
    for i in range(0, n-1):
        for j in range(i+1, n):
            if a[i] == a[j]:
                result.add(a[i])
    return result

def find_same_name_list(a):
    n = len(a)
    result = []
    for i in range(0, n-1):
        for j in range(i+1, n):
            if a[i] == a[j] and a[i] not in result:
                result.append(a[i])
    return result

name = ['Tom', 'Jerry', 'Mike', 'Tom']
print(find_same_name_set(name))
print(find_same_name_list(name)
      )
name2 = ['Tom', 'Jerry', 'Mike', 'Tom', 'Mike']
print(find_same_name_set(name2))
print(find_same_name_list(name2))
```

**Input**
+ 5 1 22 22 22 3
+ 4 98 76 20 76
+ 6 19 19 35 86 86 86
+ 1 7
+ 0

**Output**
- 1 22 3 $
- 98 76 20 76 $

- 7 $

```python
def rm_consecutive_dup(a):
    if not a:
        return []
    
    result = [a[0]]  # 첫 번째 원소는 항상 포함
    
    for i in range(1, len(a)):
        if a[i] != a[i-1]:  # 이전 원소와 다르면 추가
            result.append(a[i])
    
    return result

while True:
    line = input().split()
    n = int(line[0])
    if n == 0:
        break
    
    numbers = [int(x) for x in line[1:]]
    result = rm_consecutive_dup(numbers)
    
    # 공백으로 구분하여 출력하고 마지막에 $ 추가
    print(' '.join(map(str, result)) + ' $')
```

```python
# 단계별 테스트
def rm_consecutive_dup_debug(a):
    print(f"입력: {a}")
    if not a:
        return []
    
    result = [a[0]]
    print(f"초기화: result = {result}")
    
    for i in range(1, len(a)):
        print(f"i={i}: a[{i}]={a[i]}, a[{i-1}]={a[i-1]}", end=" ")
        if a[i] != a[i-1]:
            result.append(a[i])
            print(f"→ 다름 → result = {result}")
        else:
            print(f"→ 같음 → 추가하지 않음")
    
    return result

# 예시 테스트
numbers = [1, 22, 22, 22, 3]
result = rm_consecutive_dup_debug(numbers)
print(f"\n최종 결과: {' '.join(map(str, result))} $")
```

```python
# 잘못된 알고리즘 단계별 분석
def rm_dup_debug(a):
    print(f"\n잘못된 알고리즘 분석 - 입력: {a}")
    n = len(a)
    result = []
    
    for i in range(0, n-1):  # 마지막 원소는 확인하지 않음
        print(f"i={i}: a[{i}]={a[i]}", end=" ")
        found_duplicate = False
        for j in range(i+1, n):
            print(f"j={j}(a[{j}]={a[j]})", end=" ")
            if a[i] == a[j]:
                print(f"→ 중복 발견! break")
                found_duplicate = True
                break
        if not found_duplicate:
            result.append(a[i])
            print(f"→ 중복 없음 → result = {result}")
    
    result.append('$')
    return result

# 같은 예시로 테스트
numbers = [1, 22, 22, 22, 3]
wrong_result = rm_dup_debug(numbers)
print(f"\n잘못된 알고리즘 결과: {wrong_result}")
```
