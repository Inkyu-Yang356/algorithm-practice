```python
# BFS (Breadth-First Search) 구현
def print_all_friends(g, start):
    qu = [start]
    done = set()
    done.add(start)
    while qu:
        v = qu.pop(0)
        print(v, end=' ')
        for w in g[v]:
            if w not in done:
                qu.append(w)
                done.add(w)
                
print_all_friends(fr_info, 'Summer'); print()
print_all_friends(fr_info, 'Tom')
```

```python
# BFS (Breadth-First Search) 구현 (거리 정보 추가)
def print_all_friends(g, start):
    qu = [(start, 0)]
    done = set()
    done.add(start)
    while qu:
        v, d = qu.pop(0)
        print(v, d)
        for w in g[v]:
            if w not in done:
                qu.append((w, d+1))
                done.add(w)
                
print_all_friends(fr_info, 'Summer'); print()
print_all_friends(fr_info, 'Tom')
```

```python
# DFS (Depth-First Search) 구현
def print_all_friends_dfs(g, start):
    st = [start]
    done = set()
    done.add(start)
    while st:
        v = st.pop()
        print(v, end=' ')
        for w in g[v]:
            if w not in done:
                st.append(w)
                done.add(w)

print_all_friends_dfs(fr_info, 'Summer'); print()
print_all_friends_dfs(fr_info, 'Tom')
```

```python
from collections import deque
li1 = [1, 1, 1, 1, 1]
li2 = [4, 1, 2, 1]

# 리스트의 원소에 + 또는 - 부호를 붙여서 타겟 넘버를 만드는 방법의 수 구하기
# Input : [1,1,1,1,1]
# Output : 5  (1+1+1+1-1 = 3 을 만드는 방법의 수)
def target_number(li, target):
    qu = deque([(0, 0)])  # (현재 인덱스, 현재 합계)
    count = 0
    while qu:
        idx, total = qu.popleft()
        if idx == len(li):
            if total == target:
                count += 1
        else:
            qu.append((idx + 1, total + li[idx]))
            qu.append((idx + 1, total - li[idx]))
    return count

print(target_number(li1, 3))  # Output: 5
print(target_number(li2, 4))  # Output: 2
```

```python
def solution(numbers, target):
    if not numbers and target == 0 :
        return 1
    elif not numbers:
        return 0
    else:
        return solution(numbers[1:], target-numbers[0]) + solution(numbers[1:], target+numbers[0])
    
print(solution([1,1,1,1,1], 3)) # Output: 5
print(solution([4,1,2,1], 4))   # Output: 2
```
