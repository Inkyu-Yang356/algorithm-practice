### Palindrome

- 낱말 사이에 있는 공백이나 문장 기호 등은 무시.
- mom
- wow
- noon
- level
- radar
- kayak
- racecar
- God's dog
- Madam, I'm Adam.
- "Was it a car or a cat I saw" = "Was it a car or a cat I saw"

```python
import re

s = input()

s = s.strip()
s = re.sub(r'[^a-zA-Z0-9]', '', s).lower()

qu, st = list(), list()
for c in s:
    qu.append(c)
    st.append(c)


def is_palindrome(qu, st):
    while qu and st:
        if qu.pop(0) != st.pop():
            return False
    return True

print(is_palindrome(qu, st))
```

```python
def palindrome(s):
    # 큐와 스택을 리스트로 정의
    qu = []
    st = []
    for x in s: # 1단계: 문자열의 알파벳 문자를 각각 큐와 스택에 넣음
        if x.isalpha(): # 해당 문자가 알파벳이면 큐와 스택에 각각 그 문자를 추가
            qu.append(x.lower())
            st.append(x.lower())
    
    while qu and st: # 2단계: 큐와 스택에 들어 있는 문자를 꺼내면서 비교, 큐에 문자가 남아 있는 동안 반복
        if qu.pop(0) != st.pop(): # 큐와 스택에서 꺼낸 문자가 다르면 Palindrome이 아님
            return False # 다르면 회문이 아님
    return True

print(palindrome('Wow'))
print(palindrome('Madam, I\'m Adam'))
print(palindrome('Madam, I am Adam'))
```

```python
# How .isalpha() works  
x = 'Madam, I\'m Adam'
y = 'Madam, I am Adam'
s = [x, y]
li = []
for i in s:  # Each String in the list
    for j in i:  # Each character in the String
        print(j, j.isalpha())  # Check if the character is alphabetic
        if j.isalpha():
            li.append(j.lower())
print(li)
print("".join(li))
```

### Dictionary Set

- Input : ex) name = ["Tom", "Jerry", "Mike", "Tom"]
- Output : ex) ["Tom"]

```python
name_dict = {}
li = input().split()
for name in li:
    if name_dict.get(name):
        name_dict[name] += 1
    else:
        name_dict[name] = 1

print(name_dict, '\n')  # Mid-Check

results = []
for name in name_dict:
    if name_dict[name] >= 2:
        results.append(name)

print(results)
```

```python
def find_same_name(a):
    # 1단계: 각 이름의 등장 횟수를 딕셔너리로 만듦
    name_dict = {}
    for name in a:  # 리스트 a에 있는 자료들을 차례로 반복
        if name in name_dict:  # 이름이 name_dict에 이미 있으면
            name_dict[name] += 1  # 등장 횟수를 1 증가
        else:  # 새 이름이면
            name_dict[name] = 1  # 등장 횟수를 1로 저장
    # 2단계: 만들어진 딕셔너리에서 등장 횟수가 2 이상인 것을 결과에 추가
    result = set()  # 결과를 저장할 빈 집합
    for name in name_dict:
        if name_dict[name] >= 2:
            result.add(name)
            
    return result

name1 = ["Tom", "Jerry", "Mike", "Tom"]
print(find_same_name(name1))

name2 = ["Tom", "Jerry", "Mike", "Tom", "Jerry", "Tom"]
print(find_same_name(name2))
```

```python
sample_info = {
    39: "Justin",
    14: "John",
    67: "Mike", 
    105: "Summer"
}

def get_name(dict, number):
    if number in dict:
        return dict[number]
    else:
        return "?"
    
print(get_name(sample_info, 105))  # Output: Summer
print(get_name(sample_info, 777))  # Output: ?
```

### Practice 2
- Remove consecutively duplicate numbers

| arr | answer |
|-----|--------|
| [1,1,3,3,0,1,1] | [1,3,0,1] |
| [4,4,4,3,3] | [4,3] |

```python
def solution(arr):
    answer = []
    for i in arr:
        if i not in answer[len(answer)-1:]:
            answer.append(i)
    return answer

print(solution([1,1,3,3,0,1,1]))  # Output: [1,3,0,1]
print(solution([4,4,4,3,3]))      # Output: [4,3]
```
